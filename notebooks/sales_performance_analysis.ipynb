## Sales Trends Over Time

Analyzing sales trends over time helps detect seasonality, growth patterns,
or irregular fluctuations that may impact forecasting and planning.


The following visualization highlights key patterns and supports
interpretation of sales performance across products.


```python
import matplotlib.pyplot as plt

# Count products
product_counts = df["Product"].value_counts()

# Bar chart
product_counts.plot(kind="bar")

plt.title("Distribution of Products")
plt.xlabel("Product")
plt.ylabel("Number of Records")
plt.tight_layout()
plt.show()
```

### üîç Visualization Insight

The bar chart confirms that product categories are evenly distributed,
with no single product dominating the dataset.


### üîç Key Insight
A small number of products account for a large share of records, indicating a skewed distribution.
This suggests focusing analysis and decisions on these top products.
