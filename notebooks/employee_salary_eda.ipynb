## ðŸ“ˆ Salary Distribution
This visualization highlights the overall salary distribution across employees.
It helps identify salary concentration, spread, and potential outliers that may
represent senior roles or high-value positions.

```python
sns.histplot(df["salary"], kde=True)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.show()

```

## ðŸ“Š Salary by Experience Level
To better understand compensation progression, employees were grouped into
experience levels based on years of experience:

- Entry (0â€“2 years)
- Mid (3â€“5 years)
- Senior (6â€“10 years)
- Lead (10+ years)

This allows for clearer comparison of salary trends across career stages.

```python
sns.scatterplot(x="years_experience", y="salary", data=df)
plt.title("Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## ðŸ¢ Salary by Department
Department-level analysis reveals how compensation varies across business units.
This can highlight departments with higher average pay and greater salary dispersion.

```python
sns.boxplot(x="department", y="salary", data=df)
plt.title("Salary Distribution by Department")
plt.xticks(rotation=45)
plt.show()
```

## Key Findings / Insights
- Salaries generally increase with years of experience
- Significant salary variation exists across departments
- Senior and Lead employees exhibit wider salary dispersion,
  suggesting role-based differences or negotiation effects
- Certain departments consistently offer higher average compensation
